```python
import pandas as pd
import numpy as np

import sklearn
from sklearn.neighbors import NearestNeighbors

all_genres = set()
```

```python
movies = pd.read_csv('ml-1m/movies.dat', sep='::', names=['movie_id', 'title', 'genre'], engine='python')
```

```python
movies.head()
```

```python
for index, row in movies.iterrows():
    genres = row['genre'].split("|")
    for genre in genres:
        all_genres.add(genre)
```

```python
print(len(all_genres))
```

```python
indices = movies.title.unique()
```

```python
movie_genres = []
movie_genres = all_genres
movie_genre = pd.DataFrame(index= indices, columns=movie_genres)
```

```python
for index, row in movies.iterrows():
    genres = row['genre'].split("|")
    for genre in genres:
        movie_genre.set_value(row['title'], genre, 1)
```

```python
movie_genre.fillna(0, inplace=True)
movie_genre.head(10)
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(movie_genre, test_size = 0.2)
print(train.shape)
print(test.shape)
```

```python
knn = NearestNeighbors(n_neighbors=10).fit(train)
```

```python
from scipy.spatial.distance import jaccard as mk
errors = []
def calculate_error():
    for row in range(len(test)):
        e = 0
        test_point = test.iloc[row].values.reshape((1,18))
        _x, _y = knn.kneighbors(test_point)
        for i in range(10):
            v = train.iloc[_y[0][i]].values.reshape((1,18))
            e += mk(test_point,v)
        e = np.sqrt(e/10)
        errors.append(e)
```

```python
calculate_error()
```

```python
print("Average RMSE of the errors on test set : ",sum(errors)/len(errors))
```
