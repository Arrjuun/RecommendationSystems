```python
import pandas as pd
import numpy as np

import sklearn
from sklearn.neighbors import NearestNeighbors

all_genres = set()
```

```python
movies = pd.read_csv('ml-1m/movies.dat', sep='::', names=['movie_id', 'title', 'genre'], engine='python')
```

```python
movies.head()
```

```python
for index, row in movies.iterrows():
    genres = row['genre'].split("|")
    for genre in genres:
        all_genres.add(genre)
```

```python
print(len(all_genres))
```

```python
indices = movies.title.unique()
```

```python
movie_genres = []
movie_genres = all_genres
movie_genre = pd.DataFrame(index= indices, columns=movie_genres)
```

```python
for index, row in movies.iterrows():
    genres = row['genre'].split("|")
    for genre in genres:
        movie_genre.set_value(row['title'], genre, 1)
```

```python
movie_genre.fillna(0, inplace=True)
movie_genre.head(10)
```

```python
knn = NearestNeighbors(n_neighbors=10).fit(movie_genre)
```

```python
test_point = movie_genre.loc["Toy Story (1995)"].values.reshape((1,18))
re_x, re_y = knn.kneighbors(test_point)
results = re_y[0]
```
